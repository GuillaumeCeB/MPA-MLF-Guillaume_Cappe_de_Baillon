## Exercise 1.3

Write a script that will convert an arbitrary integer to a binary number (the number will be represented as a string with only *0* and *1*). Avoid using the *bin()* function or any other pythons default functions

```python
def int_to_binary(num: int) -> str:
    if num == 0:
        return "0"
    
    binary_str = ""
    is_negative = num < 0
    num = abs(num)
    
    while num > 0:
        binary_str = str(num % 2) + binary_str
        num = num // 2
    
    if is_negative:
        binary_str = "-" + binary_str
    
    return binary_str

# Example usage
print(int_to_binary(10))
print(int_to_binary(-10))
print(int_to_binary(0))
```

## Exercise 1.4 - The Fibonacci Sequence:

The Fibonacci Sequence is a series of numbers. The following number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13. The following number in this series above is 8 + 13 = 21


Your task is to implement a function *fibonacci* that takes an integer as an input and returns a list that contains all Fibonacci numbers with values lower than the input integer

```python
def fibonacci(upper_threshold: int) -> list:
    if upper_threshold <= 0:
        return []
        
    # First two Fibonacci numbers
    result = [0]
    if upper_threshold > 1:
        result.append(1)
        
    # Keep adding next Fibonacci number until we reach threshold
    while result[-1] + result[-2] < upper_threshold:
        result.append(result[-1] + result[-2])
    
    return result

# Example usage
print(fibonacci(10))  # Output: [0, 1, 1, 2, 3, 5, 8]
```

#### Advanced

Try to implement the function using multiple approaches: *Iterative approach*, *Recusrion*, *Memoization*

```python
# Iterative approach
def fibonacci_iterative(upper_threshold: int) -> list:
    if upper_threshold <= 0:
        return []
    
    # Initialize with first two Fibonacci numbers
    fib_list = [0]
    if upper_threshold > 1:
        fib_list.append(1)
    
    # Generate Fibonacci numbers until reaching threshold
    while True:
        next_num = fib_list[-1] + fib_list[-2]
        if next_num >= upper_threshold:
            break
        fib_list.append(next_num)
    
    return fib_list

# Recursive approach
def fibonacci_recursive(upper_threshold: int) -> list:
    def fib_recur(n):
        if n <= 1:
            return n
        return fib_recur(n-1) + fib_recur(n-2)
    
    result = []
    i = 0
    while True:
        fib_num = fib_recur(i)
        if fib_num >= upper_threshold:
            break
        result.append(fib_num)
        i += 1
    return result

# Memoization approach
def fibonacci_memoization(upper_threshold: int) -> list:
    memo = {}
    
    def fib_memo(n):
        if n in memo:
            return memo[n]
        if n <= 1:
            return n
        
        memo[n] = fib_memo(n-1) + fib_memo(n-2)
        return memo[n]
    
    result = []
    i = 0
    while True:
        fib_num = fib_memo(i)
        if fib_num >= upper_threshold:
            break
        result.append(fib_num)
        i += 1
    return result

print(fibonacci_iterative(10))
print(fibonacci_recursive(10))
print(fibonacci_memoization(10))
```

```python
import random

def rock_paper_scissors() -> None:
    # Valid moves
    valid_moves = ['rock', 'paper', 'scissors']
    
    # Get computer's move
    computer_move = random.choice(valid_moves)
    
    # Get user's move
    user_move = input("Enter your move (rock/paper/scissors): ").lower()
    
    # Validate user input
    if user_move not in valid_moves:
        print("Invalid move! Please enter rock, paper, or scissors.")
        return
    
    # Print moves
    print(f"Your move: {user_move}")
    print(f"Computer's move: {computer_move}")
    
    # Determine winner
    if user_move == computer_move:
        print("It is a tie")
    elif (
        (user_move == 'rock' and computer_move == 'scissors') or
        (user_move == 'paper' and computer_move == 'rock') or
        (user_move == 'scissors' and computer_move == 'paper')
    ):
        print("You win")
    else:
        print("You lose")

# Example usage
rock_paper_scissors()
```

#### Advanced

Extend the Rock, Paper, Scissors game to be able to play *n* rounds. 

Notes: 
- The scrips will contain two counters - user score and computer score
- After each run, the script prints the current scores and rounds.
- When the number of rounds is reached, the script prints the result ('You lose', 'You win', 'It is a tie')

```python
import random

def rock_paper_scissors_advanced(rounds: int) -> None:
    # Valid moves
    valid_moves = ['rock', 'paper', 'scissors']
    
    # Initialize scores
    user_score = 0
    computer_score = 0
    
    # Play specified number of rounds
    for current_round in range(1, rounds + 1):
        print(f"\nRound {current_round}/{rounds}")
        print(f"Current score - You: {user_score}, Computer: {computer_score}")
        
        # Get computer's move
        computer_move = random.choice(valid_moves)
        
        # Get user's move
        user_move = input("Enter your move (rock/paper/scissors): ").lower()
        
        # Validate user input
        if user_move not in valid_moves:
            print("Invalid move! Please enter rock, paper, or scissors.")
            continue
        
        # Print moves
        print(f"Your move: {user_move}")
        print(f"Computer's move: {computer_move}")
        
        # Determine winner of the round
        if user_move == computer_move:
            print("Round tied!")
        elif (
            (user_move == 'rock' and computer_move == 'scissors') or
            (user_move == 'paper' and computer_move == 'rock') or
            (user_move == 'scissors' and computer_move == 'paper')
        ):
            print("You win this round!")
            user_score += 1
        else:
            print("Computer wins this round!")
            computer_score += 1
    
    # Print final results
    print("\nGame Over!")
    print(f"Final score - You: {user_score}, Computer: {computer_score}")
    
    # Determine overall winner
    if user_score > computer_score:
        print("You win!")
    elif computer_score > user_score:
        print("You lose!")
    else:
        print("It is a tie!")

# Example usage
rock_paper_scissors_advanced(3)  # Play 3 rounds
```

## Exercise 2.1


In this exercise, you will work with NumPy arrays and learn how to efficiently modify and process them using both loops and vectorized operations. Your task is to:

1. Implement the function *create_array_nxn* that generates an n×n NumPy array filled with numbers from n² - 1 down to 0
2. Using a loop-based approach (for-loop, while-loop), implement the function *apply_threshold_loop* that takes any NumPy array and replaces all numbers below a user-defined threshold with 0.
3. Implement function *apply_threshold_vectorized* that performs the same operation as *apply_threshold_loop*, but without loops, using NumPy vectorized operations
4. Implement function *compare_performance* that compares the execution time of functions *apply_threshold_vectorized* and *apply_threshold_loop* with **time** library and print the results


**Additional Notes**:

- Ensure n is a positive integer (n > 0).
- Use the time library to measure execution times for performance comparison.
- The function *compare_performance(n, threshold)* should:
    - Generate an n×n array.
    - Apply both loop-based and vectorized thresholding.
    - Print execution times.

```python
import numpy as np
import time
```

```python
def create_array_nxn(n: int) -> np.ndarray:
    """Creates an n×n array filled with numbers from n²-1 down to 0"""
    if n <= 0:
        raise ValueError("n must be positive")
    return np.arange(n*n-1, -1, -1).reshape((n, n))
```

```python
def apply_threshold_loop(arr: np.ndarray, threshold: int) -> np.ndarray:
    """Applies threshold using loops"""
    result = arr.copy()  # Create a copy to avoid modifying original array
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            if result[i, j] < threshold:
                result[i, j] = 0
    return result
```

```python
def apply_threshold_vectorized(arr: np.ndarray, threshold: int) -> np.ndarray:
    """Applies threshold using vectorized operations"""
    result = arr.copy()  # Create a copy to avoid modifying original array
    result[result < threshold] = 0
    return result
```

```python
def compare_performance(n: int, threshold: int) -> None:
    """Compares performance between loop and vectorized approaches"""
    # Create test array
    arr = create_array_nxn(n)
    
    # Test loop-based approach
    start_time = time.time()
    loop_result = apply_threshold_loop(arr, threshold)
    loop_time = time.time() - start_time
    
    # Test vectorized approach
    start_time = time.time()
    vectorized_result = apply_threshold_vectorized(arr, threshold)
    vectorized_time = time.time() - start_time
    
    # Print results
    print(f"Array size: {n}x{n}")
    print(f"Threshold: {threshold}")
    print(f"Loop-based time: {loop_time:.6f} seconds")
    print(f"Vectorized time: {vectorized_time:.6f} seconds")
    
    # Calculate speedup with check for zero
    if vectorized_time > 0:
        print(f"Speedup: {loop_time/vectorized_time:.2f}x")
    else:
        print("Speedup: infinite (vectorized time too small to measure)")
    
    # Verify results are identical
    if np.array_equal(loop_result, vectorized_result):
        print("Results are identical")
    else:
        print("Warning: Results differ!")

# Example usage
if __name__ == "__main__":
    # Create and display a small array
    arr = create_array_nxn(4)
    print("Original 4x4 array:")
    print(arr)
    
    # Apply threshold and show results
    threshold = 10
    print(f"\nArray after applying threshold {threshold}:")
    print(apply_threshold_vectorized(arr, threshold))
    
    # Compare performance with larger array
    print("\nPerformance comparison:")
    compare_performance(100, 5000)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def show_in_digi(input_integer: int) -> None:
    
    # Convert input integer to string
    num_str = str(input_integer)
    
    # Create list to store digit arrays
    digit_arrays = []
    
    # Convert each digit to its corresponding array
    for digit in num_str:
        if digit in numbs:
            digit_arrays.append(numbs[digit])
    
    # Add small spacing between digits
    spacing = np.zeros((5, 1))
    
    # Combine all digits with spacing
    display = digit_arrays[0]
    for digit_array in digit_arrays[1:]:
        display = np.concatenate((display, spacing, digit_array), axis=1)
    
    # Display the result using matplotlib
    plt.figure(figsize=(len(digit_arrays), 2))
    plt.imshow(display, cmap='binary')
    plt.axis('off')
    plt.show()

# Example usage
show_in_digi(5289)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)

# 1. Check what dataset.describe() does
print("\nDataset Description:")
print(df.describe())

# 2. Display rows where AveBedrms > 2 (using a more reasonable threshold)
print("\nRows where average bedrooms > 2:")
print(df[df['AveBedrms'] > 2])

# 3. Drop first and last row
df = df.iloc[1:-1]

# 4. Plot AveOccup (average occupancy) with mean line
occupancy_mean = df['AveOccup'].mean()
plt.figure(figsize=(10, 6))
plt.scatter(range(len(df)), df['AveOccup'], c='blue', alpha=0.5, label='Average Occupancy')
plt.axhline(y=occupancy_mean, color='r', linestyle='-', label=f'Mean ({occupancy_mean:.2f})')
plt.xlabel('ID')
plt.ylabel('Average Occupancy')
plt.title('Average Occupancy Distribution with Mean')
plt.legend()
plt.show()

# 5. Check for NaN values and replace with mean if any exist
print("\nColumns with NaN values:")
print(df.isnull().sum())
# Replace NaN values with column means if they exist
df = df.fillna(df.mean())

# 6. Plot latitude vs longitude
plt.figure(figsize=(10, 8))
plt.scatter(df['Latitude'], df['Longitude'], alpha=0.5)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('California Housing Locations')
plt.show()

# 7. Normalize two arbitrary columns (MedInc and AveRooms)
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())

columns_to_normalize = ['MedInc', 'AveRooms']
normalized_df = df.copy()
for col in columns_to_normalize:
    normalized_df[f'{col}_normalized'] = min_max_normalize(df[col])

print("\nOriginal and Normalized values for selected columns:")
print(normalized_df[columns_to_normalize + [f'{col}_normalized' for col in columns_to_normalize]].head())

# 8. Create correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', aspect='auto')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
